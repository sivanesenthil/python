```python
import seaborn as sns
import pandas as pd

# Load the iris dataset
iris = sns.load_dataset('iris')

# Display all records
iris
```

```python
# Count the number of records
record_count = iris.shape[0]
record_count
```

```python
# Find the distinct Species in the dataset
distinct_species = iris['species'].unique()
distinct_species
```

```python
# Get the average sepal length for each Species
avg_sepal = iris.groupby('species')['sepal_length'].mean()
avg_sepal
```

```python
# Find the minimum and maximum petal length for each Species
min_max_petal = iris.groupby('species')['petal_length'].agg(['min', 'max'])
min_max_petal
```

```python
# Retrieve the records where the sepal width is greater than 3.0
sepal_width = iris[iris['sepal_width'] > 3.0]
sepal_width
```

```python
# Find the Species with the highest average petal width
highest_petal = iris.groupby('species')['petal_width'].mean().idxmax()
highest_petal
```

```python
# List the top 5 records with the largest sepal length
largest_sepal_length = iris.nlargest(5, 'sepal_length')
largest_sepal_length
```

```python
# Find the total number of records for each Species
total_records_per_species = iris['species'].value_counts()
total_records_per_species
```

```python
# Get the avg, min, max values for sepal length and petal length for each Species
summary_stats = iris.groupby('species').agg({
    'sepal_length': ['mean', 'min', 'max'],
    'petal_length': ['mean', 'min', 'max']
})
summary_stats
```
